```python
import json
import jsonlines 

def load_jsonl(name):
    v = []
    with open(name) as f:
        for line in f:
            v.append(json.loads(line))
    print("No. of examples", len(v))
    return v
```

```python
vvv = load_jsonl(folder + 'generations.txt.jsonl')
```

```python
f = jsonlines.open(folder + 'generations_1.txt.jsonl', "w")
with jsonlines.open(folder + 'generations.txt.jsonl') as reader:
    for line_number, line in enumerate(reader):
        line['rhyming_schema'] = ['RHYME_' + i for i in line['rhyming_schema']]
        f.write(line)
        if (line_number + 1) % 50000 == 0:
            print(str(line_number + 1) + ' lines have done.')
```
